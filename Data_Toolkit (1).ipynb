15. How does NumPy handle multidimensional arrays?

ans-NumPy's primary object, the `ndarray` (n-dimensional array), is specifically designed to efficiently handle multidimensional arrays. This is a core feature that makes NumPy so powerful for numerical computing.

NumPy handles multidimensional arrays:

- **`ndarray` Object:** The `ndarray` is a homogeneous data structure, meaning all elements within the array must be of the same data type. This allows NumPy to store and manipulate the data in a contiguous block of memory, which significantly improves performance.
- **Shape and Dimensions:** An `ndarray` has a `shape` attribute, which is a tuple of integers indicating the size of the array along each dimension. The number of dimensions is called the rank or `ndim`.
- **Indexing and Slicing:** NumPy provides powerful and flexible indexing and slicing capabilities to access and manipulate elements or subarrays within a multidimensional array. You can use integers, slices, boolean arrays, or fancy indexing to select data.
- **Vectorized Operations:** As mentioned earlier, NumPy allows for vectorized operations on entire arrays, regardless of their dimensions. This means you can perform arithmetic, logical, and other operations on multidimensional arrays without explicit loops, leading to much faster execution.
- **Broadcasting:** NumPy's broadcasting rules allow you to perform operations on arrays with different shapes, as long as their dimensions are compatible according to the broadcasting rules. This is particularly useful when working with multidimensional arrays and scalars or arrays of different shapes.
- **Memory Layout:** NumPy arrays can have different memory layouts (C-order or Fortran-order), which can impact performance depending on how you access and process the data.

17. Explain the difference between apply() and map() in Pandas?

ans- In Pandas, both apply() and map() are used to apply a function to a Series or DataFrame, but they have different use cases and behaviors:

map():

1. Primarily used on a Pandas Series.
2. Used for element-wise transformation or substitution of values in a Series.
3. Can take a dictionary, a Series, or a function as an argument.
4. If a dictionary or Series is used, it performs a lookup and replaces values.
5. If a function is used, it applies the function to each element in the Series.
6 Generally faster than apply() for element-wise operations on a Series.

apply():

1. Can be used on both Series and DataFrames.
2. On a Series, it behaves similarly to map() for element-wise operations, but map() is often preferred for performance.
3. On a DataFrame, apply() is used to apply a function along an axis (rows or columns).
4. It can apply a function to each row (axis=1) or each column (axis=0).
The function passed to apply() should be able to handle a Series (when applied to rows or columns of a DataFrame) or a single value (when applied element-wise to a Series).
5. More flexible than map() as it can operate on entire rows or columns of a DataFrame.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

ans- NumPy's array slicing is significantly faster than Python's list slicing primarily due to the underlying implementation and memory management:

Homogeneous Data Type and Contiguous Memory: NumPy arrays are homogeneous, meaning all elements are of the same data type. They are also typically stored in contiguous blocks of memory. This allows NumPy to quickly calculate the memory address of any element based on its index. When you slice a NumPy array, it can directly access the required block of memory without needing to iterate through elements.

Heterogeneous Data Type and Scattered Memory: Python lists, on the other hand, can store elements of different data types. This means they cannot store elements in a contiguous block of memory in the same way as NumPy arrays. Instead, a Python list stores pointers to the memory locations of its elements, which can be scattered throughout memory.

Overhead of Object Creation: When you slice a Python list, a new list object is created to hold the sliced elements. This involves allocating memory for the new list and copying the references to the elements. This object creation and copying process adds overhead.

Views vs. Copies: In many slicing operations, NumPy returns a "view" of the original array rather than a complete copy. A view is essentially a new array object that refers to the same data in memory as the original array. This avoids the overhead of creating a new array and copying data. While some NumPy slicing operations do create copies, many common ones create views, contributing to performance. Python list slicing always creates a new list object (a copy).

C Implementation: NumPy's core operations, including slicing, are implemented in C. This low-level implementation allows for greater performance optimization compared to Python's higher-level operations.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


data = {'col1': np.random.rand(50),
        'col2': np.random.rand(50) * 100,
        'col3': np.random.randint(0, 100, 50),
        'col4': np.random.randn(50)}
df = pd.DataFrame(data)

print("Original DataFrame:")
display(df.head())


correlation_matrix = df.corr()

print("\nCorrelation Matrix:")
display(correlation_matrix)


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
